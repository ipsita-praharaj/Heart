```python
import numpy as np
from sklearn import linear_model,datasets,tree
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import keras.backend as K

config = K.tf.ConfigProto()
config.gpu_options.allow_growth = True
session = K.tf.Session(config=config)
```

```python
iri = pd.read_excel('1_YEAR_10.FOLD.xlsx')
```

```python
iris=iri #.select_dtypes(['number'])
from sklearn import preprocessing
from sklearn import model_selection
for column in iris.columns:
    if iris[column].dtype == type(object):
        le = preprocessing.LabelEncoder()
        iris[column] = le.fit_transform(iris[column])
iri2=iris
X=iri2.drop(labels=['GSTATUS'], axis=1).values
print(X.shape)
Y = iri2.GSTATUS.values
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, test_size=0.2, random_state=10)

X_train, X_val, Y_train, Y_val = model_selection.train_test_split(X_train,Y_train, test_size=0.2, random_state=10)
Y_test2=Y_test
from keras.utils import to_categorical
Y_val = to_categorical( Y_val)
Y_train = to_categorical(Y_train)
Y_test = to_categorical(Y_test)
from keras.models import Sequential 
from keras.layers import Dense, Activation 
output_dim = nb_classes = 2
model = Sequential() 
model.add(Dense(output_dim, input_dim=116, activation='softmax')) 
batch_size = 128 
nb_epoch = 20
model.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy']) 
history = model.fit(X_train, Y_train, batch_size=batch_size, epochs=nb_epoch,verbose=1, validation_data=(X_val, Y_val)) 
score = model.evaluate(X_test, Y_test, verbose=0) 
print('Test score:', score[0]) 
print('Test accuracy:', score[1])
```

```python
test_set_predictions = [model.predict(X_test[i].reshape((1,116))) for i in range(X_test.shape[0])]
print(test_set_predictions)
test_misclassification_percentage = 0
for i in range(len(test_set_predictions)):
    if test_set_predictions[i]!=Y_test2[i]:
        test_misclassification_percentage+=1
test_misclassification_percentage *= 100/len(Y_test2)
print ('test misclassification percentage =', test_misclassification_percentage, '%')
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
test_set_predictions=np.array([k[0] for k in test_set_predictions])
print(test_set_predictions.shape)
Y_test2=Y_test2.reshape(1852, 1)
print(accuracy_score(Y_test2, test_set_predictions))
print(confusion_matrix(Y_test2, test_set_predictions))

import sklearn.metrics as metrics
fpr, tpr, thresholds = metrics.roc_curve(Y_test2, test_set_predictions)
plt.plot(fpr, tpr)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.title('ROC curve for diabetes classifier')
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.grid(True)
def evaluate_threshold(threshold):
    print('Sensitivity:', tpr[thresholds > threshold][-1])
    print('Specificity:', 1 - fpr[thresholds > threshold][-1])
evaluate_threshold(0.5)
print(metrics.roc_auc_score(Y_test2, test_set_predictions))
```
